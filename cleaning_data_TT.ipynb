```python
income = pd.read_csv("income.csv",engine='python')
income.head()
```

```python
income_order = ["State_Name", "State_ab", "City", "ALand", "AWater", "Median", "Stdev"]
income = income[income_order]
income.head()
```

```python
grp = income.groupby('City')

mean = grp.mean()[['ALand','AWater']]
maximum = grp.max()[['State_Name','State_ab','Median','Stdev']]

new_income = maximum.merge(mean,left_index=True,right_index=True)
new_income.head()
```

```python
new_income.shape
```

```python
new_income.isnull().sum()
```

```python
new_income.info()
```

```python
# Remove 0+, +0i, i from Stdev
new_income['Stdev'] = new_income['Stdev'].map(lambda x: x.lstrip('0+').rstrip('+0i''i'))

# Change Stdev to float 
new_income['Stdev'] = pd.to_numeric(new_income['Stdev'])

# Round off the value of Stdev to 2 decimal places 
new_income.round({"Stdev":2})
```

```python
new_income.info()
```

```python
new_income.isnull().sum()
```

```python
# Drop the null columns where all values are null
new_income =new_income.dropna(axis='columns', how='all')

# Drop the null rows
new_income = new_income.dropna()
```

```python
new_income.isnull().sum()
```

```python
new_income.info()
```

```python
new_income.to_csv("income_cleaned.csv", index=False)
```
