```python
income = pd.read_csv("income.csv",engine='python')
income.head()
```

```python
income.rename(columns={'Median': 'St_Median'}, inplace=True)
```

```python
income
```

```python
income_order = ["State_Name", "State_ab", "City", "ALand", "AWater", "St_Median", "Stdev"]
income = income[income_order]
income.head()
```

```python
#income = income.groupby('City').mean().reset_index()
income.groupby([“City”], as_index=False).agg({“State_Name”: “first”, “State_ab”: “first”, “ALand”: “mean”, “AWater”: “mean”, “St_Median”: “mean”, “Stdev”: “mean”})
```

```python
#income.groupby(['State_Name','State_ab','City', 'ALand', 'AWater', 'Median', 'Stdev']).mean().reset_index()
#income
```

```python
grp = income.groupby('City')

mean = grp.mean()[['ALand','AWater']]
maximum = grp.mean()[['State_Name','State_ab','Median','Stdev']]

income = maximum.merge(mean,left_index=True,right_index=True)
income.head()
```

```python
income.shape
```

```python
income.isnull().sum()
```

```python
income.info()
```

```python
# Remove 0+, +0i, i from Stdev
income['Stdev'] = income['Stdev'].map(lambda x: x.lstrip('0+').rstrip('+0i''i'))

# Change Stdev to float 
income['Stdev'] = pd.to_numeric(income['Stdev'])

# Round off the value of Stdev to 2 decimal places 
income.round({"Stdev":2})
```

```python
income.info()
```
